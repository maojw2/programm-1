```python
import numpy as np
import matplotlib.pyplot as plt


class MultiLayerPerceptron:
    def __init__(self, layer_sizes, activations, learning_rates):
        
        
        self.layer_sizes = layer_sizes
        self.activations = activations
        self.learning_rates = learning_rates
        self.weights = [np.random.uniform(-2.0, 2.0, (layer_sizes[i], layer_sizes[i-1])) for i in range(1, len(layer_sizes))]
        self.biases = [np.random.uniform(-2.0, 2.0, (size, 1)) for size in layer_sizes[1:]]

    def activation(self, x, func):
        if func == "tanh":
            return np.tanh(x)
        elif func == "logistic":
            return 1 / (1 + np.exp(-x))
        elif func == "identity":
            return x

    def activation_derivative(self, x, func):
        if func == "tanh":
            return 1 - np.tanh(x)**2
        elif func == "logistic":
            return self.activation(x, "logistic") * (1 - self.activation(x, "logistic"))
        elif func == "identity":
            return np.ones_like(x)

    def forward(self, x):
        self.a = [x]
        self.z = []
        for i in range(len(self.layer_sizes) - 1):
            z = np.dot(self.weights[i], self.a[-1]) + self.biases[i]
            a = self.activation(z, self.activations[i])
            self.z.append(z)
            self.a.append(a)
        return self.a[-1]

    def backward(self, y):
        deltas = [None] * (len(self.layer_sizes) - 1)
        deltas[-1] = (self.a[-1] - y) * self.activation_derivative(self.z[-1], self.activations[-1])
        
        for i in range(len(deltas) - 2, -1, -1):
            deltas[i] = np.dot(self.weights[i+1].T, deltas[i+1]) * self.activation_derivative(self.z[i], self.activations[i])
        
        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rates[i] * np.dot(deltas[i], self.a[i].T)
            self.biases[i] -= self.learning_rates[i] * deltas[i]

    def train(self, inputs, targets, epochs, filename="learning_curve.txt"):
        with open(filename, "w") as file:
            for epoch in range(epochs):
                total_error = 0
                for x, y in zip(inputs, targets):
                    x, y = x.reshape(-1, 1), y.reshape(-1, 1)
                    self.forward(x)
                    self.backward(y)
                    total_error += np.sum((self.a[-1] - y) ** 2)
                avg_error = total_error / len(inputs)
                file.write(f"{epoch + 1}\t{avg_error}\n")
                print(f"Epoch {epoch + 1}, Error: {avg_error}")

    def evaluate(self, inputs, targets):
        correct_predictions = 0
        for x, y in zip(inputs, targets):
            x = x.reshape(-1, 1)
            prediction = self.forward(x)
            if np.argmax(prediction) == np.argmax(y):
                correct_predictions += 1
        accuracy = correct_predictions / len(inputs)
        print(f"Test Accuracy: {accuracy * 100:.2f}%")
        return accuracy


def load_data(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()[2:]  # skip the first two header lines
        inputs, targets = [], []
        for line in lines:
            values = list(map(float, line.split()))
            inputs.append(values[:2])  # first two values are inputs (N=2)
            targets.append(values[2:])  # last value is the target (M=1)
    return np.array(inputs), np.array(targets)

# Define the structure of the MLP and training parameters
layer_sizes = [2, 5, 3, 1]          # 4 layers total
activations = ["tanh", "logistic", "identity"]  # 3 activations
learning_rates = [0.1, 0.05, 0.01]  # 3 learning rates

# Initialize MLP
mlp = MultiLayerPerceptron(layer_sizes, activations, learning_rates)


# Load data and start training
train_inputs, train_targets = load_data(r"C:\Users\12951\Downloads\PA-A_training_data_02.txt")
test_inputs, test_targets = load_data(r"C:\Users\12951\Downloads\PA-A_training_data_02.txt")

mlp.train(train_inputs, train_targets, epochs=500)
mlp.evaluate(test_inputs, test_targets)

# Plotting the learning curve
data = np.loadtxt("learning_curve.txt")
plt.plot(data[:, 0], data[:, 1])
plt.xlabel("Epoch")
plt.ylabel("Average Error")
plt.title("Learning Curve")
plt.show()
```
